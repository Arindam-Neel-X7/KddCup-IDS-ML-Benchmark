```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
df = pd.read_csv('/content/drive/MyDrive/KddCup/data/kdd_clean.csv')

df.head()
```

```python
df = pd.get_dummies(df, columns=['protocol_type', 'service', 'flag'])
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
def run_gbc(test_size):

    print(f"===== TEST SIZE: {test_size} =====\n")

    # Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        random_state=42,
        stratify=y
    )

    # Scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    # PCA
    pca = PCA(n_components=10)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)

    # Model
    model = GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=3,
        random_state=42
    )

    model.fit(X_train, y_train)

    # Prediction
    y_pred = model.predict(X_test)

    # Evaluation
    print("Accuracy:", accuracy_score(y_test, y_pred))

    print("\nClassification Report:\n")
    print(classification_report(y_test, y_pred))

    print("\nConfusion Matrix:\n")
    print(confusion_matrix(y_test, y_pred))
    print("\n")

    # Plot
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f"GBC (Test Size = {test_size})")
    plt.show()
```

**Gradient Boosting Classifier with 80,20**

```python
run_gbc(0.2)
```

**Gradient Boosting Classifier with 60,40**

```python
run_gbc(0.4)
```

**Gradient Boosting Classifier with 40,60**

```python
run_gbc(0.6)
```
